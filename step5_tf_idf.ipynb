# An√°lise TF IDF

```python
import pandas as pd 
import numpy as np
```

```python
df_all = pd.read_csv('tfidf/df_tf_idf_max_all.csv', sep='\t')
df_positivo = pd.read_csv('tfidf/df_tf_idf_max_positive.csv', sep='\t')
df_negativo = pd.read_csv('tfidf/df_tf_idf_max_negative.csv', sep='\t')
df_neutro = pd.read_csv('tfidf/df_tf_idf_max_neutral.csv', sep='\t')
```

```python
df_positivo.head()
```

```python
df_all_top_50 = df_all.sort_values(by=['values'], ascending=False).head(50)
df_positivo_top_50 = df_positivo.sort_values(by=['values'], ascending=False).head(50)
df_negativo_top_50 = df_negativo.sort_values(by=['values'], ascending=False).head(50)
df_neutro_top_50 = df_neutro.sort_values(by=['values'], ascending=False).head(50)
```

```python
df_all_top_50.sort_values(by=['values'], ascending=True).plot(x='names', y='values', kind='barh', figsize=(5,15), title='Todos os artigos')
```

```python
df_positivo_top_50.sort_values(by=['values'], ascending=True).plot(x='names', y='values', kind='barh', figsize=(5,15), title='artigos que influenciam positivamente')
```

```python
df_negativo_top_50.sort_values(by=['values'], ascending=True).plot(x='names', y='values', kind='barh', figsize=(5,15), title='artigos que influenciam negativamente')
```

```python
df_neutro_top_50.sort_values(by=['values'], ascending=True).plot(x='names', y='values', kind='barh', figsize=(5,15), title='artigos que n√£o influenciam')
```
